```python
fp=open('Alice.txt','r',encoding='UTF-8')
line=fp.readline()
dic={}
import matplotlib.pyplot as plt
import numpy as np
while line:
    s=line.split()
    for x in s:
        if x not in dic:
            dic[x]=1
            
        else:
            dic[x]=dic[x]+1
    line=fp.readline()
fp.close()
num=[]
for key in dic:
    num.append(dic[key])
print(len(num))
num.sort()
num.reverse()

from scipy import optimize

def test_func(x,amp,alpha):
    return amp*x**alpha

a,a_co=optimize.curve_fit(test_func,range(1,len(num)+1),num)
print(a)
print(a_co)
plt.plot(range(1,len(num)+1),test_func(range(1,len(num)+1),a[0],a[1]),'r')   
plt.loglog(range(1,len(num)+1),num)
plt.xlabel('Rank of the word')
plt.ylabel('Number of the occurance')
plt.title("Zipf's law")
plt.show()
```
