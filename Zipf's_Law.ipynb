```python
fp=open('Alice.txt','r',encoding='UTF-8')
line=fp.readline()
dic={}
import matplotlib.pyplot as plt
import numpy as np
while line:
    s=line.split()
    for x in s:
        if x not in dic:
            dic[x]=1
            
        else:
            dic[x]=dic[x]+1
    line=fp.readline()
fp.close()
num=[]
for key in dic:
    num.append(dic[key])
print(len(num))
num.sort()
num.reverse()
    
plt.loglog(range(len(num)),num)
plt.xlabel('Rank of the word')
plt.ylabel('Number of the occurance')
plt.title("Zipf's law")
plt.show()
```
